# INTRODUCTION

Using the most popular unsupervised machine learning, we are going to predict, depending on its value, in which group they will be.

# LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# CREATING THE DATASET

We are going to generate our own dataset, this dataset is going to be of 1000 entries, two dimensions (1000,2), four groups, a standard variation of 1.8 and a random state that could be any number.

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples = 1000, n_features = 2, centers = 4, cluster_std = 1.8, random_state = 101)
```

# LET'S PLOT OUR DATASET

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap= "rainbow")
```

# UNSUPERVISED MACHINE LEARNING MODEL

First we need to import the correct module of scikit learn.

Then we are going to build our model with four clusters, it could be less or more but because of our dataset is built in four centers, it is better this way.

Last, we are going to train it.

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 4)
kmeans.fit(data[0])
```

```python
kmeans.labels_
```

# LET'S PLOT THE MACHINE LEARNING MODEL

We are going to draw two viz, the first one is the ML prediction, when we can see the four groups clearly. The second one is the original dataset.
Pd. The colors are random.

```python
fig, (ax1,ax2) = plt.subplots( 1,2, sharey = True, figsize = (10,6))

ax1.set_title("K Means")
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap = "rainbow")

ax2.set_title("Original")
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap = "rainbow")
```
